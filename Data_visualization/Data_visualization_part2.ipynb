```python
## Scatter Plot with trendlines
import plotly.express as px
iris=px.data.iris()

px.scatter(iris,x='sepal_width',y='sepal_length',color='species',trendline='ols')
```

```python
tip=px.data.tips()
px.bar(tip,x='day',y='total_bill',color="smoker",barmode='group')
```

```python
px.bar(tip,x='day',y='total_bill',color="sex",barmode='group',facet_row='time',facet_col='day',category_orders={'day':["Thur", "Fri", "Sat", "Sun"],"time": ["Lunch", "Dinner"]})
```

```python
## Animations.
import plotly.express as px

df = px.data.gapminder()
fig = px.scatter(df, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="iso_num", color="continent", hover_name="country", facet_col="continent",
           log_x=True, size_max=45, range_x=[100,100000], range_y=[25,90])
fig.show()
```
