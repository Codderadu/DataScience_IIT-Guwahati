# Seaborn

Seaborn updates Matplotlib's (run command (rc) parameters) to improve aesthetics
Seaborn sits on the top of matplot lib code & provides better functionality ie better syntax, better defaults
Seaborn runs with matplotlib
Seaborn also integrates with pandas -- It allows seaborn group, aggregate and plot pandas dataframes

```python
import seaborn as sns
```

```python
df=sns.load_dataset("mpg")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
sns.relplot(data=df,x='model_year',y='mpg',col='origin',hue='cylinders')
```

A density plot is a representation of the distribution of a numeric variable. It uses a kernel density estimate to show the probability density function of the variable (see more). It is a smoothed version of the histogram and is used in the same concept.

Density Plot is a type of data visualization tool. It is a variation of the histogram that uses 'kernel smoothing' while plotting the values. It is a continuous and smooth version of a histogram inferred from a data. ... Density plots can be made using pandas, seaborn, etc.

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable.

```python
sns.kdeplot(df['horsepower'],shade=True)
#Most Cars have the Horse power of 80, some cars have around 150
```

```python
# Create KDE plot with custom shade color
sns.kdeplot(df.horsepower, shade=True, color='red')

# Set plot labels and title
plt.xlabel('Horsepower')
plt.ylabel('Density')
plt.title('KDE Plot')

# Display the plot
plt.show()
```

```python
#Cumulative Density Plot (cdf)
sns.kdeplot(df.horsepower,cumulative=True)
```
